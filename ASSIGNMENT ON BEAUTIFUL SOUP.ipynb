```python
URL = "https://presidentofindia.nic.in/former-presidents"
```

```python
import requests
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content , 'html5lib')
print(soup.prettify())
```

```python
list_of_presidents = soup.find_all('div', class_="desc-sec")
```

```python
print(list_of_presidents)
```

```python
president_names = []
terms_of_office = []
for president_info in list_of_presidents:
        name = president_info.find('h3').text.strip()
        term_of_office = president_info.find('h5').text.strip()
        president_names.append(name)
        terms_of_office.append(term_of_office)
```

```python
data = {'Name of former president': president_names , 'Term of Office' :terms_of_office}
```

```python
df = pd.DataFrame(data)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

URL = 'https://www.cnbc.com/world/?region=world'
r = requests.get(URL)
soup = BeautifulSoup(r.text, 'html5lib')

data = []
news_items = soup.find_all('li', class_='LatestNews-item')
    
for item in news_items:
    time = item.find('time', class_='LatestNews-timestamp').text
    Headline = item.find('a', class_='LatestNews-headline').text
    Newslink = item.find('a', class_='LatestNews-headline')['href']
    data.append({'Time': time, 'Headline': Headline, 'NewsLink': Newslink})

df_1 = pd.DataFrame(data)
print(df_1)
```

```python
URL = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"


r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')

titles = []
authors = []
dates = []
links = []

articles = soup.select('ul.sc-9zxyh7-0 li.sc-9zxyh7-1')

for article in articles:
    Paper_Title = article.select_one('h2').text.strip()
    Authors = article.select_one('.sc-1w3fpd7-0').text.strip()
    Published_date = article.select_one('.sc-1thf9ly-2 span').text.strip()
    Paper_URL = article.select_one('a')['href']

    titles.append(Paper_Title)
    authors.append(Authors)
    dates.append(Published_date)
    links.append(Paper_URL)


data = {
    'Paper_Title': titles,
    'Authors': authors,
    'Published_dates': dates,
    'Paper_URL': links
}

df = pd.DataFrame(data)
print(df)
```

```python
URL = "https://www.dineout.co.in/delhi-restaurants/buffet-special"

response = requests.get(URL)

soup = BeautifulSoup(response.text, 'html5lib')


image_elements = soup.find_all('img', class_='no-img')
image_links = [img['data-src'] for img in image_elements if 'data-src' in img.attrs]

        
df_image_links = pd.DataFrame({'Image Links': image_links})

    
restaurant_sections = soup.find_all('div', class_='restaurant-info-section')

        
restaurant_names = []
cuisine_lists = []
locations = []
ratings = []

        
for restaurant_section in restaurant_sections:
    restaurant_name_element = restaurant_section.find('a', class_='restnt-name')
    restaurant_name = restaurant_name_element.text.strip()
    restaurant_names.append(restaurant_name)

            
    cuisine_element = restaurant_section.find('span', class_='double-line-ellipsis')
    cuisine_elements = cuisine_element.find_all('a', href=True) 
    cuisine_names = [cuisine.text.strip() for cuisine in cuisine_elements]
    cuisine_lists.append(cuisine_names)

            
    location_div = restaurant_section.find('div', class_='restnt-loc ellipsis')
    locality_element = location_div.find('a', {'data-type': 'LocalityClick'}) 
    area_element = location_div.find('a', {'data-type': 'AreaClick'}) 
    locality = locality_element.text.strip() 
    area = area_element.text.strip() 
    location = f"{locality}, {area}" 
    locations.append(location)

            
    rating_element = restaurant_section.find('div', class_='restnt-rating')
    restaurant_rating = rating_element.text.strip()
    ratings.append(restaurant_rating)

        
data_restaurants = {
            'Restaurant Name': restaurant_names,
            'Cuisine Names': cuisine_lists,
            'Location': locations,
            'Restaurant Rating': ratings,
        }
df_restaurants = pd.DataFrame(data_restaurants)

        
result_df = pd.concat([df_restaurants, df_image_links], axis=1)

        
pd.set_option('display.max_colwidth', None)
print(result_df)
```
